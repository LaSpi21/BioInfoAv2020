```
# Ejecericio 1 c
#Funcion traducir porque la voy a necesitar
def traducir(secuencia, devolver_codones = False):
    proteina = ''
    codones = []
    n=0
    diccionario_adn_proteina = {diccionario_protein_adn[i][j]: i for i in diccionario_protein_adn.keys() for j in range(len(diccionario_protein_adn[i]))}
    while secuencia[n:n+3] != '' and diccionario_adn_proteina[secuencia[n:n+3]] != 'STOP':
        codones.append(secuencia[n:n+3])
        proteina += diccionario_adn_proteina[secuencia[n:n+3]][0]
        n+=3
    if devolver_codones:
        return proteina, codones
    return proteina
```

```
# Ejercicio 1 d (opcional)
""" tenemos el problema de que varios ADN distintos pueden dar la misma proteina
Asi que planteamos 3 formas disintas de resolverlo."""
#i) Crear todos los ADN que pueden codificar una dada proteina..
#Contra: el numero de ADNs que pueden generar una dada proteina crece muy rapido con el largo de la proteina..
def Posibles_ADN(Proteina, imprime = False):
    """
    Crea todos los posibles ADNs que pueden generar una proteina
    Usese bajo su propio riesgo..."""
    import itertools
    lista_de_posibles_codones = []
    for aa in Proteina:
        lista_de_posibles_codones.append(diccionario_protein_adn[aa])
    lista_de_posibles_codones_con_posicion = []
    for i in range(0, len(lista_de_posibles_codones)):
        for j in lista_de_posibles_codones[i]:
            lista_de_posibles_codones_con_posicion.append((i,j))
    tempdict = {}
    for elemento in lista_de_posibles_codones_con_posicion:
        id = elemento[0]
        if id in tempdict:
            tempdict[id].append(elemento)
        else:
            tempdict[id] = [elemento]
    lista_de_tuplas = list(itertools.product(*list(tempdict.values())))
    lista_de_ADNs = []
    for tupla in lista_de_tuplas:
        ADN = ''
        for pos,codon in tupla:
            ADN += codon
        lista_de_ADNs.append(ADN)
        if imprime == True:
            print(ADN)
    return lista_de_ADNs
#Creamos una prot aleatoria chiquita para que no se nos vaya de las manos..
Largo = 10
Proteina_aleatoria = 'M'
for i in range(Largo-1):
    Proteina_aleatoria += random.choice(Aminoacidos)
print(Proteina_aleatoria)

M = Posibles_ADN(Proteina_aleatoria)
len(M)
```

```
#Ejercicio 1 d (opcional) continuacion
import numpy.random as nr

Largo = 1200
Proteina_aleatoria = 'M'
for i in range(Largo-1):
    Proteina_aleatoria += nr.choice(Aminoacidos, p = [Diccionario_de_frecuencias_vertebrados[i] for i in Aminoacidos])
print(Proteina_aleatoria)

def Generador_Posibles_ADN(Proteina, imprime = False):
    """
    Devuelve un generador que genera todos los posibles ADNs que
    resultan en una proteina dada.
    Cada iteracion devuelve un posible ADN al azar que codifica la 
    proteina dada.
    Esta version es computacionalmente mas amigable que
    Posibles_ADN() y se lleva bien con proteinas de cualquier largo"""
    import itertools
    lista_de_posibles_codones = []
    for aa in Proteina:
        lista_de_posibles_codones.append(diccionario_protein_adn[aa])
    lista_de_posibles_codones_con_posicion = []
    for i in range(0, len(lista_de_posibles_codones)):
        for j in lista_de_posibles_codones[i]:
            lista_de_posibles_codones_con_posicion.append((i,j))
    tempdict = {}
    for elemento in lista_de_posibles_codones_con_posicion:
        id = elemento[0]
        if id in tempdict:
            tempdict[id].append(elemento)
        else:
            tempdict[id] = [elemento]
    for e in tempdict.values():
        random.shuffle(e)
    return (''.join([codon for pos,codon in i]) for i in itertools.product(*list(tempdict.values())))

M = Generador_Posibles_ADN(Proteina_aleatoria)

#%%
next(M)
```
